# **Basic Descriptive Statistics on Dataset:**
## **1. Load a Small Dataset:**

We use the Iris dataset.

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("iris")
# Show first rows
print(" First 5 rows:")
print(df.head(), "\n")

```

## **2. Calculate Descriptive Statistics:**
#### a) **Full descriptive summary using describe()**

```python
df.describe()
# Descriptive statistics
print(" Descriptive Statistics (describe()):")
print(df.describe(), "\n")
```

####  **b) Calculate Individual Statistics Manually:**

```python
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
min_values = df.min(numeric_only=True)
max_values = df.max(numeric_only=True)
std_values = df.std(numeric_only=True)

# Manual statistics
print(" Mean:\n", df.mean(numeric_only=True), "\n")
print(" Median:\n", df.median(numeric_only=True), "\n")
print(" Min:\n", df.min(numeric_only=True), "\n")
print(" Max:\n", df.max(numeric_only=True), "\n")
print(" Standard Deviation:\n", df.std(numeric_only=True), "\n")

mean_values, median_values, min_values, max_values, std_values
```

## **3. Check Missing Values:**

```python
# Missing values check
print("ðŸ”¹ Missing Values:")
print(df.isnull().sum())
```
